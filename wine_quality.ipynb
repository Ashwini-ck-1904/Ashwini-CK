```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

# GET THE DATA

```python
df = pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/winequalityN.csv')
```

# UNDERSTANDING DATASET 

```python
df.info()
```

```python
df
```

# DATA CLEANING

```python
df.info()
```

# column	Non-Null Count	Missing Values
# fixed acidity	6487	10
# volatile acidity	6489	8
# citric acid	6494	3
# residual sugar	6495	2
# chlorides	6495	2
# pH	6488	9
# sulphates	6493	4

```python
df['fixed acidity'].fillna(df['fixed acidity'].mean(), inplace=True)
df['volatile acidity'].fillna(df['volatile acidity'].mean(), inplace=True)
df['citric acid'].fillna(df['citric acid'].mean(), inplace=True)
df['residual sugar'].fillna(df['residual sugar'].mean(), inplace=True)
df['chlorides'].fillna(df['chlorides'].mean(), inplace=True)
df['pH'].fillna(df['pH'].mean(), inplace=True)
df['sulphates'].fillna(df['sulphates'].mean(), inplace=True)
```

```python
df.info()
```

# DATA MANIPULATION


```python
df['sugar_acidity_ratio'] = df['residual sugar'] / (df['fixed acidity'] + 1e-5)
```

```python
print(df['type'].unique())
```

```python
df['type'] = df['type'].map({'white': 0, 'red': 1})
```

```python
df.info()
```

# White wines (type 0):

>>Are sweeter (higher residual sugar)

>>Are more acidic (lower pH)

>>Use more sulfur dioxide for preservation

>>Have slightly higher average quality

>>Have a higher sugar-acidity ratio â†’ may taste smoother or fruitier

# Red wines (type 1):

>>Are more acidic in structure (higher fixed acidity)

>>Have more volatile acidity (vinegar-like sharpness)

>>Have higher sulphates and chlorides

>>Tend to be denser and more structured

```python
#if one type or quality score dominates
df['type'].value_counts()
df['quality'].value_counts()
```

Most wines are rated 5 or 6 , these categories make upto 77% of all the samples.

Very few wines are rated 3 or 9, 
which means:

The dataset is imbalanced

Extreme quality scores (3, 9) might not provide enough data for accurate modeling.

```python
#Understand how variables relate to wine type or quality.

#1)Average alcohol by quality
df.groupby('quality')['alcohol'].mean()

```

As quality increases, the average alcohol content  increases,

>>Higher quality wines tend to have more alcohol.

>>alcohol could be positively correlated with quality.

```python
#which features are related to quality
df.corr()['quality'].sort_values(ascending=True)
```

>>Positively Correlated (Higher values â†’ better quality):
Among all the positively corealated features, alcohol stands out as the strongest positive predictor of wine quality.

>>Negatively Correlated (Higher values â†’ lower quality):
Among all the negetively corelated features, ensity and volatile acidity are the most negatively associated with quality, meaning higher values of these may indicate poorer quality.



```python
#Calculate sugar acidity ratio compare new insights.
df.groupby('quality')['sugar_acidity_ratio'].mean()
```

# DATA VISUALIZATION

```python
# quality distribution
plt.figure(figsize=(5,3))
a=sns.countplot(x='quality',data=df, palette='coolwarm')
a.bar_label(a.containers[0])
plt.title('Wine Quality Distribution')
plt.xlabel('Quality Score')
plt.ylabel('Count')

```

```python
plt.figure(figsize=(5,3))
b=sns.countplot(x='type', data=df, palette='Set2')
b.bar_label(b.containers[0])
plt.title('Count of Wine Types')
plt.xlabel('Type')
plt.ylabel('Count')


```

```python
plt.figure(figsize=(5,3))
c=sns.boxplot(x='quality', y='alcohol', data=df, palette='viridis')
plt.title('Alcohol Content by Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Alcohol (%)')

```

```python
plt.figure(figsize=(5,3))
d=sns.boxplot(x='quality', y='density', data=df, palette='rocket')
plt.title('Wine Density by Quality')
plt.xlabel('Quality')
plt.ylabel('Density')

```

```python
plt.figure(figsize=(8,4))
e=sns.scatterplot(x='alcohol', y='density', hue='quality', data=df, palette='coolwarm')
plt.title('Alcohol vs. Density (Colored by Quality)')
plt.show()
```
